```python
#code snippet 1
import pandas as pd
import numpy as np
# Specify the path to your CSV file
url = 'https://docs.google.com/spreadsheets/d/1XSBE7VwTJ-xB300r2JAwZesaAbGTLvMgS4qRxXfDjys/pub?gid=0&single=true&output=csv'
# Read the CSV file into a pandas DataFrame
data = pd.read_csv(url)
data.shape
# Display the first few rows of the DataFrame to verify
# display(data.head())
```

```python
# Code snippet 3
# Display the data columns
data.columns
```

```python
#derive age from birthdate
from datetime import datetime
today = datetime.now()
# Convert 'Birth Date' to datetime objects with explicit format
data['Birth Date'] = pd.to_datetime(data['Birth Date'], format='%m/%d/%y')

# Correct birth years that are in the future due to two-digit year interpretation
# If a birth year is after the current year, subtract 100 years
data.loc[data['Birth Date'].dt.year > today.year, 'Birth Date'] -= pd.offsets.DateOffset(years=100)

# Calculate age based on current date, using only the year part
data['Age'] = today.year - data['Birth Date'].dt.year

# Filter data where 'Birth Date' is less than or equal to today and make an explicit copy
dataa = data[data['Birth Date'] <= today].copy()
dataa
```

```python
#code snippet 3.2
#separate City-ZipCode-State into 3 varaibles:
dataa[['City', 'ZipCode', 'State']]=dataa['City-ZipCode-State'].str.split('-', expand=True)
```

Predictive Analysis
Target Variable ='Avg Monthly Spend'

```python
# Code snippet 4
# Predictive analysis of Target1: Avg Monthly Spend
# Setting the target variable
y = 'Avg Monthly Spend'
```

```python
# Code snippet 5
# Setting the ignored variables list (noise/redundant/feeder and variables that are more than 20% data)
ignored_cols = ['Customer ID', 'First Name', 'Last Name', 'Street Address1','Title',
                 'Middle Name', 'Suffix', 'Street Address2','Birth Date','City', 'State','City-ZipCode-State','eBook Subscriber Flag']
```

```python
# Code snippet 6
# Setting the categorical variables
cat_cols = ['Education Level', 'Occupation', 'Gender', 'Marital Status', 'Home Owner Status', 'ZipCode']
```

```python
# Code snippet 7
# Setting the numerical columns
num_cols = ['Number of Cars Owned', 'Number of Children At Home',
            'Total Number of Children', 'Annual Income', 'Age']
```

```python
# Code snippet 8
# Importing regression modules from pycaret
from pycaret.regression import *
```

```python
# Code snippet 9
# Setting/configuring the ML experiment
regression_setup = setup(dataa,
                        target=y,
                        ignore_features=ignored_cols,
                        categorical_features=cat_cols,
                         numeric_features=num_cols,
                        train_size=0.8)
```

```python
# Code snippet 10
# Invoking ML algorithms
compare_models()
```

```python
# Code snippet 11
# Building an ML model with the best algo from the leaderboard above
best_model = create_model('lightgbm')
```

```python
# Code snippet for tuning the best model
tuned_model = tune_model(best_model)
```

```python
# Code snippet 12
# Predicting the standard scores using the best model
predicted_AvgMonthlySpend_best = predict_model(best_model, dataa)

# Exporting the predictions to a csv file
predicted_AvgMonthlySpend_best.to_csv('eBook Store Avg Monthly Spend Prediction.csv')
```

```python
# Predicting the standard scores using the tuned model
predicted_AvgMonthlySpend_tuned = predict_model(tuned_model, dataa)

# Exporting the predictions to a csv file
predicted_AvgMonthlySpend_tuned.to_csv('eBook Store Avg Monthly Spend Prediction tuned.csv')
```

```python
# Code snippet 13
# Displaying the prediction scores
predicted_AvgMonthlySpend_best
```

```python
#Code snippet 15
# Display the original 'Avg Monthly Spend', the prediction from the best model, and the prediction from the tuned model
comparison_df = dataa[['Avg Monthly Spend']].copy()
comparison_df['Prediction (Best Model)'] = predicted_AvgMonthlySpend_best['prediction_label']
comparison_df['Prediction (Tuned Model)'] = predicted_AvgMonthlySpend_tuned['prediction_label']

display(comparison_df.head())
```
